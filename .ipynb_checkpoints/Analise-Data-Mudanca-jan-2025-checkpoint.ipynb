# O tempo de mudança adiciona informação em relação ao momento de compra?

## Objetivo da análise e resumo da metodologia

O objetivo dessa análise é entender se vale a pena usar "Data de Mudança" junto com momento de  compra para ver se conseguimos melhorar a qualificação dos leads usando as variáveis em conjunto. Separadamente, ambas explicam o maior avanço no funil. 

Para definir se faz sentido usar as variáveis em conjunto compararei
- Avanço para pré análise de crédito
- Avanço para análise de crédito
- transferência de vendas para ops

Usando o teste Z adaptado para proporções, usando a mesma metodologia aplicada nessa análise https://www.notion.so/moreamora/An-lise-Momento-de-compra-dcf6acefecdf4615879034d107e640a3

## Exploração dos dados

### gráfico mostrando a evolução do preenchimento da variável data de mudança

```python
import pandas as pd
import matplotlib.pyplot as plt

# Verificar valores únicos na coluna "Data de Mudança"
print(df['Data de Mudança'].unique())

# Substituir valores nulos em "Data de Mudança" por 'Não preenchida'
df['Data de Mudança'] = df['Data de Mudança'].fillna('Não preenchida')

# Converter "Data de criação" para datetime, se necessário
if not pd.api.types.is_datetime64_any_dtype(df['Data de criação']):
    df['Data de criação'] = pd.to_datetime(df['Data de criação'])

# Criar a coluna de mês/ano a partir de "Data de criação"
df['AnoMes'] = df['Data de criação'].dt.to_period('M')

# Agrupar os dados por mês/ano de criação e "Data de Mudança"
grouped_data = df.groupby(['AnoMes', 'Data de Mudança']).size().unstack(fill_value=0)

# Criar o gráfico de barras empilhadas
grouped_data.plot(kind='bar', stacked=True, figsize=(12, 6))

# Personalizar o gráfico
plt.title('Distribuição de "Data de Mudança" por Mês de Criação')
plt.xlabel('Mês e Ano de Criação')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.legend(title='Data de Mudança')
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

### Tabela com totais de leads aprovados por categoria e grupo - leads sem data de mudança preenchida foram filtrados. 

```python
import pandas as pd

# Filtrar apenas os valores preenchidos na coluna "Data de Mudança" e status de cadastro "Aprovado"
df_filt = df[(df['Data de Mudança'] != 'Não preenchida') & (df['status_cadastro'] == 'Aprovado')]

# Criar a tabela cruzada com "Momento de Compra" nas linhas e "Data de Mudança" nas colunas
pivot_table = pd.crosstab(df_filt['Momento de Compra'], df_filt['Data de Mudança'], margins=True, margins_name='Total')

# Definir a nova ordem para "Momento de Compra" (linhas)
ordem_momento_compra = [
    "Só quero saber como funciona",
    "Estou no começo da jornada",
    "Estou visitando algumas opções",
    "Quero comprar e já tenho imóvel definido",
    "Total"  # Certifique-se de que "Total" permaneça no final
]

# Definir a nova ordem para "Data de Mudança" (colunas)
ordem_data_mudanca = [
    "Imediatamente",
    "1 a 2 meses",
    "3 a 4 meses",
    "Acima de 4 meses",
    "Não tenho certeza ainda",
    "Total"  # Certifique-se de que "Total" permaneça no final
]

# Reordenar as linhas e colunas
pivot_table = pivot_table.reindex(index=ordem_momento_compra, columns=ordem_data_mudanca)

# Exibir a tabela
print(pivot_table)
```

## percentual da coluna

```python
# Calcular percentuais por coluna (excluindo a coluna "Total")
pivot_table_percent_col = pivot_table.div(pivot_table.loc['Total'], axis=1) * 100

# Exibir a tabela como porcentagens por coluna
print(pivot_table_percent_col)
```

```python
Existe uma tendência a leads com maior urgência de mudança terem um momento de compra. 
```

# Análise 1- comparativo de pré análise

## % de avanços Observados - Pré Análise

```python
# Criar coluna indicando leads que avançaram para pré-análise (usando .loc para evitar warnings)
df_filt.loc[:, 'Avancou_Pre_Analise'] = ~df_filt["Date entered \"Pré-análise crédito (Funil.Vendas)\""].isna()

# Calcular a tabela cruzada com percentuais
pivot_table_pre_analise = pd.crosstab(
    df_filt['Momento de Compra'], 
    df_filt['Data de Mudança'], 
    values=df_filt['Avancou_Pre_Analise'], 
    aggfunc='mean', 
    margins=True, 
    margins_name='Total'
) * 100  # Converter para porcentagem

# Reordenar as linhas e colunas na ordem desejada
pivot_table_pre_analise = pivot_table_pre_analise.reindex(index=ordem_momento_compra, columns=ordem_data_mudanca)

# Exibir a tabela formatada com porcentagens
pivot_table_pre_analise_styled = pivot_table_pre_analise.style.format("{:.1f}%")
pivot_table_pre_analise_styled
```

## Número absoluto de avanços para pre-analise por grupo

```python
# Calcular a tabela cruzada com valores absolutos (quantidade de leads que avançaram)
pivot_table_absolute = pd.crosstab(
    df_filt['Momento de Compra'], 
    df_filt['Data de Mudança'], 
    values=df_filt['Avancou_Pre_Analise'], 
    aggfunc='sum', 
    margins=True, 
    margins_name='Total'
).astype(int)  # Converter os valores para inteiros

# Reordenar as linhas e colunas na ordem desejada
pivot_table_absolute = pivot_table_absolute.reindex(index=ordem_momento_compra, columns=ordem_data_mudanca)

# Exibir a tabela
pivot_table_absolute_styled=pivot_table_absolute.style.format("{:.0f}")
pivot_table_absolute_styled
```

## Valores esperados para avanço para pré análise

Abaixo os valores estão multiplicando o % de avanço para pré an´lise por momento de compra (total da linha da tabela  % de avanços observados pelo número de leads.

```python
# Usar os percentuais de avanço para cada linha (convertendo de % para decimal)
percentual_avanco = pivot_table_pre_analise["Total"] / 100  # Convertendo de % para decimal

# Multiplicar o percentual de avanço pelo número de leads observados em cada célula
valores_esperados = pivot_table.mul(percentual_avanco, axis=0)

# Aplicar formatação para exibir os valores sem casas decimais
valores_esperados_styled = valores_esperados.style.format("{:.0f}")

# Exibir a tabela formatada
valores_esperados_styled
```

##  Valor Z para pré análise

```python
import numpy as np

# Probabilidade total da linha (P), expandida para ter as mesmas dimensões de O e E
P = pivot_table_pre_analise["Total"] / 100  # Convertendo para decimal
P = P.values[:, np.newaxis]  # Transformar em matriz (coluna expandida para múltiplas colunas)

# Valores observados (O)
O = pivot_table_absolute

# Valores esperados (E)
E = valores_esperados

# Calcular o valor Z para cada célula
Z = (O - E) / np.sqrt(E * (1 - P))

# Função para aplicar destaque condicional
def highlight_significant(value):
    # Destacar valores maiores que 1.96 ou menores que -1.96
    if value > 1.96 or value < -1.96:
        return 'background-color: yellow'
    else:
        return ''  # Sem destaque

# Aplicar o estilo de highlight às células usando Styler.map
Z_styled = Z.style.format("{:.2f}").map(highlight_significant)

# Exibir a tabela formatada com destaque
Z_styled

```

## Tabela de % de avanço para pré análise por grupo 

```python
# Criar a mesma máscara de destaque para os valores Z
def create_highlight_mask(Z):
    return Z.map(lambda x: 'background-color: yellow' if x > 1.96 or x < -1.96 else '')

# Gerar a máscara de destaque a partir da tabela Z
highlight_mask = create_highlight_mask(Z)

# Aplicar a máscara na tabela pivot_table_pre_analise
pivot_table_pre_analise_styled = pivot_table_pre_analise.style.format("{:.1f}%").apply(lambda x: highlight_mask, axis=None)

# Exibir a tabela com os mesmos highlights
pivot_table_pre_analise_styled
```

## Conclusões para pré análise

Para pré análise, não existe diferença significativa entre usar data de mudança e momento de compra em conjunto. 
Apenas a célula "Só Quero Saber como FUnciona" e Data de Mudança "3-4 meses". Não há evidencias abundantes de 
que essas variaveis em conjunto funcionem para prever melhor os leads que avançarão para pré análise

# Análise 2 - avanço para análise de crédito

## % de avanços Observados -  Análise

```python
# Criar coluna indicando leads que avançaram para pré-análise (usando .loc para evitar warnings)
df_filt.loc[:, 'Avancou_Analise'] = ~df_filt["Date entered \"Análise crédito (Funil.Vendas)\""].isna()

# Calcular a tabela cruzada com percentuais
pivot_table_analise = pd.crosstab(
    df_filt['Momento de Compra'], 
    df_filt['Data de Mudança'], 
    values=df_filt['Avancou_Analise'], 
    aggfunc='mean', 
    margins=True, 
    margins_name='Total'
) * 100  # Converter para porcentagem

# Reordenar as linhas e colunas na ordem desejada
pivot_table_analise = pivot_table_analise.reindex(index=ordem_momento_compra, columns=ordem_data_mudanca)

# Exibir a tabela formatada com porcentagens
pivot_table_analise_styled = pivot_table_analise.style.format("{:.1f}%")
pivot_table_analise_styled
```

## # absoluto de valores observados - Análise

```python
# Calcular a tabela cruzada com valores absolutos (quantidade de leads que avançaram)
pivot_table__analise_absolute = pd.crosstab(
    df_filt['Momento de Compra'], 
    df_filt['Data de Mudança'], 
    values=df_filt['Avancou_Analise'], 
    aggfunc='sum', 
    margins=True, 
    margins_name='Total'
).astype(int)  # Converter os valores para inteiros

# Reordenar as linhas e colunas na ordem desejada
pivot_table__analise_absolute = pivot_table__analise_absolute.reindex(index=ordem_momento_compra, columns=ordem_data_mudanca)

# Exibir a tabela
pivot_table__analise_absolute_styled=pivot_table__analise_absolute.style.format("{:.0f}")
pivot_table__analise_absolute_styled
```

## Valores esperados para avanço para  análise

```python
# Usar os percentuais de avanço para cada linha (convertendo de % para decimal)
percentual_avanco_1 = pivot_table_analise["Total"] / 100  # Convertendo de % para decimal

# Multiplicar o percentual de avanço pelo número de leads observados em cada célula
valores_esperados_1 = pivot_table.mul(percentual_avanco_1, axis=0)

# Aplicar formatação para exibir os valores sem casas decimais
valores_esperados_1_styled = valores_esperados_1.style.format("{:.1f}")

# Exibir a tabela formatada
valores_esperados_1_styled
```

## Valores Z para análise

```python
import numpy as np

# Probabilidade total da linha (P), expandida para ter as mesmas dimensões de O e E
P1 = pivot_table_analise["Total"] / 100  # Convertendo para decimal
P1 = P1.values[:, np.newaxis]  # Transformar em matriz (coluna expandida para múltiplas colunas)

# Valores observados (O)
O1 = pivot_table__analise_absolute

# Valores esperados (E)
E1 = valores_esperados_1

# Calcular o valor Z para cada célula
Z1 = (O1 - E1) / np.sqrt(E1 * (1 - P1))

# Função para aplicar destaque condicional
def highlight_significant(value):
    # Destacar valores maiores que 1.96 ou menores que -1.96
    if value > 1.96 or value < -1.96:
        return 'background-color: yellow'
    else:
        return ''  # Sem destaque

# Aplicar o estilo de highlight às células usando Styler.map
Z1_styled = Z1.style.format("{:.2f}").map(highlight_significant)

# Exibir a tabela formatada com destaque
Z1_styled

```

## Conclusão

Como no caso anterior, aqui também temos poucas evidências de que a Data de Mudança adicione poder explicativo para o avanço de leads para a análise de crédito.

# Analise 3 - Transferencia Vendas Ops

## % de avanço - vendas ops 

```python
# Criar coluna indicando leads que avançaram para VENDAS OPS
df_filt.loc[:, 'Avancou_vendas_ops'] = ~df_filt['Date entered "Transferência vendas <>ops (Funil.Vendas)"'].isna()

# Calcular a tabela cruzada com percentuais
pivot_table_vendas_ops = pd.crosstab(
    df_filt['Momento de Compra'], 
    df_filt['Data de Mudança'], 
    values=df_filt['Avancou_vendas_ops'], 
    aggfunc='mean', 
    margins=True, 
    margins_name='Total'
) * 100  # Converter para porcentagem

# Reordenar as linhas e colunas na ordem desejada
pivot_table_vendas_ops = pivot_table_vendas_ops.reindex(index=ordem_momento_compra, columns=ordem_data_mudanca)

# Exibir a tabela formatada com porcentagens
pivot_table_vendas_ops_styled = pivot_table_vendas_ops.style.format("{:.1f}%")
pivot_table_vendas_ops_styled
```

## Valores Observados

```python
# Calcular a tabela cruzada com valores absolutos (quantidade de leads que avançaram para Transferência Vendas Ops)
pivot_table_vendas_ops_absolute = pd.crosstab(
    df_filt['Momento de Compra'], 
    df_filt['Data de Mudança'], 
    values=df_filt['Avancou_vendas_ops'],  # Usando 'Avancou_vendas_ops'
    aggfunc='sum', 
    margins=True, 
    margins_name='Total'
).astype(int)  # Converter os valores para inteiros

# Reordenar as linhas e colunas na ordem desejada
pivot_table_vendas_ops_absolute = pivot_table_vendas_ops_absolute.reindex(index=ordem_momento_compra, columns=ordem_data_mudanca)

# Exibir a tabela formatada com valores absolutos
pivot_table_vendas_ops_absolute_styled = pivot_table_vendas_ops_absolute.style.format("{:.0f}")
pivot_table_vendas_ops_absolute_styled
```

## Valores Esperados

```python
# Usar os percentuais de avanço para cada linha (convertendo de % para decimal)
percentual_avanco_vendas_ops = pivot_table_vendas_ops["Total"] / 100  # Convertendo para decimal

# Multiplicar o percentual de avanço pelo número de leads observados em cada célula
valores_esperados_vendas_ops = pivot_table.mul(percentual_avanco_vendas_ops, axis=0)

# Aplicar formatação para exibir os valores sem casas decimais
valores_esperados_vendas_ops_styled = valores_esperados_vendas_ops.style.format("{:.1f}")

# Exibir a tabela formatada
valores_esperados_vendas_ops_styled
```

## Valores Z para a análise

```python
import numpy as np

# Probabilidade total da linha (P), expandida para ter as mesmas dimensões de O e E
P_vendas_ops = pivot_table_vendas_ops_absolute["Total"] / pivot_table_vendas_ops_absolute["Total"].sum()  # Probabilidade total da linha
P_vendas_ops = P_vendas_ops.values[:, np.newaxis]  # Transformar em matriz (coluna expandida para múltiplas colunas)

# Valores observados (O)
O_vendas_ops = pivot_table_vendas_ops_absolute

# Valores esperados (E)
E_vendas_ops = valores_esperados_vendas_ops

# Calcular o valor Z para cada célula
Z_vendas_ops = (O_vendas_ops - E_vendas_ops) / np.sqrt(E_vendas_ops * (1 - P_vendas_ops))

# Função para aplicar destaque condicional
def highlight_significant(value):
    # Destacar valores maiores que 1.96 ou menores que -1.96
    if value > 1.96 or value < -1.96:
        return 'background-color: yellow'
    else:
        return ''  # Sem destaque

# Aplicar o estilo de highlight às células usando Styler.map
Z_vendas_ops_styled = Z_vendas_ops.style.format("{:.2f}").map(highlight_significant)

# Exibir a tabela formatada com destaque
Z_vendas_ops_styled
```

## Tabela de probabilidades marcando céluLas significantes

```python
# Criar a mesma máscara de destaque para os valores Z de Vendas Ops
def create_highlight_mask(Z_vendas_ops):
    return Z_vendas_ops.map(lambda x: 'background-color: yellow' if x > 1.96 or x < -1.96 else '')

# Gerar a máscara de destaque a partir da tabela Z_vendas_ops
highlight_mask_vendas_ops = create_highlight_mask(Z_vendas_ops)

# Aplicar a máscara na tabela formatada pivot_table_vendas_ops_styled
pivot_table_vendas_ops_styled = pivot_table_vendas_ops_styled.format("{:.1f}%").apply(
    lambda _: highlight_mask_vendas_ops, axis=None
)

# Exibir a tabela formatada com os mesmos highlights
pivot_table_vendas_ops_styled
```

## Conclusões

Até agora, leads que têm data de mudança  "acima de 4 meses" não avançaram para transferência vendas ops. Até existem indícios de que existe um efeito positivo para leads com data de mudança "imediatamente", porém, exceto para o momento de compra "visitando algumas opções", ainda não chegamos num número amostral que seja estatisticamente significante. 

# Conclusões Gerais

Em pré análise de crédito e em análise de crédito, o uso de "Data de mudança" agrega pouca informação em relação ao "momento de compra". Entretanto, para "Transferência Vendas Ops" já conseguimos observar algumas distinções. Primeiro, leads com Data de Mudança maior do que 4 meses não seguem para transferência vendas ops. O tamanho amostral já é suficiente para que o efeito seja estatisticamente significante de forma independente. Depois, para leads que "estão visitando algumas opções", quem quer comprar imediatamente avança significativamente mais para vendas ops.
